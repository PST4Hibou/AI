```python
from datasets import load_from_disk
from transformers import ASTFeatureExtractor
from datasets import load_dataset
from IPython.display import Audio as DisplayAudio
from datasets import Audio

num_proc = 24
batch_size = 32
sampling_rate = 16000
feature_extractor = ASTFeatureExtractor()
```

```python
dataset = load_from_disk("../../dataset/ds_1_raw_aug_0-5_balanced_chunked.hf")
```

```python
def preprocess(batch):
    for item in batch["audio"],batch["labe"]:
        print(item)
        # print(audio)
        inputs = feature_extractor(
            item["audio"],
            sampling_rate=sampling_rate,
            padding="max_length",
            truncation=True,
            return_tensors="pt",
        )
        item["input_values"] = inputs["input_values"]
        items.append(items)

    return items
```

```python
new_ds = dataset.map(preprocess, batched=True, num_proc=num_proc, batch_size=batch_size)
```
